Handle Duplicates in dataset

```python
df_dup = data.copy()

# Duplicate the first row
row = df_dup.iloc[:1]

# Append the duplicated row using concat
df_dup = pd.concat([df_dup, row], ignore_index=True)

```

```python
df_dup[df_dup.duplicated()]
```

drop the duplicated rows in the data

```python
df_dup.drop_duplicates()
```
